```python
def plotOrb(n):
    weights = [0., 0., 0., 0.]
    weights[n] = 1
    dt = get_charge_density(['Mg'], [np.array([0., 0., 0.])], weights_list=[weights])
    a = np.zeros((100, 3))
    a[:, 1] = np.arange(0., 10., 10./len(a))
    rho = dt(a)
    plt.plot(rho)
```

```python
def sample_plane(basis, lattice_const, h, k, l, nsamples = 20):
    xmax, ymax, zmax = basis * np.array([h, k, l])
    xsample = np.concatenate(list(np.ones(nsamples + 1)*i for i in range(nsamples + 1)))/(nsamples/lattice_const)
    ysample = np.concatenate(list(np.arange(nsamples + 1) for i in range(nsamples + 1)))/(nsamples/lattice_const)
    #xsample, ysample = np.mgrid[:nsamples + 1, :nsamples + 1]/(nsamples/lattice_const)
    zsample = (-lattice_const + h * xsample + k * ysample) / l
    return xsample, ysample, zsample
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate

def plot_heatmap(x, y, z):
    # Generate data:
    #x, y, z = 10 * np.random.random((3,10))

    # Set up a regular grid of interpolation points
    xi, yi = np.linspace(x.min(), x.max(), 100), np.linspace(y.min(), y.max(), 100)
    xi, yi = np.meshgrid(xi, yi)

    # Interpolate
    rbf = scipy.interpolate.Rbf(x, y, z, function='linear')
    zi = rbf(xi, yi)

    plt.imshow(zi, vmin=z.min(), vmax=z.max(), origin='lower',
               extent=[x.min(), x.max(), y.min(), y.max()])
    #plt.scatter(x, y, c=z)
    #plt.colorbar()
    plt.show()
```
